```python
import pandas as pd
import numpy as np
# pd.set_option('display.max_rows', 5)
# pd.set_option('display.max_columns', 20)
```

```python
df1 = pd.DataFrame({'Column1': range(100), 'Column2': range(100), 'Column3': range(100)})
df1['Column1'] = df1['Column1'].astype('str').str.zfill(2)
df1['Column2'] = df1['Column2'].astype('str').str.zfill(2)
df1['Column3'] = df1['Column3'].astype('str').str.zfill(2)
combine = []
for x in df1['Column1']:
    for y in df1['Column2']:
        for z in df1['Column3']:
            if int(x) - int(y) != 0 and int(x) - int(z) != 0 and int(y) - int(z) != 0:
                t = [x,y,z]
                t.sort()
                if t not in combine:
                    combine.append(t)
data = pd.DataFrame(combine)
data.to_excel('Data.xlsx', sheet_name = 'combine_3')
```
