```python
import pandas as pd
df1 = pd.DataFrame({'Column1': range(100),
                    'Column2': range(100)})
df1['Column1'] = df1['Column1'].astype('str').str.zfill(2)
df1['Column2'] = df1['Column2'].astype('str').str.zfill(2)
combine = []
for x in df1['Column1']:
    for y in df1['Column2']:
        v = [x,y]
        v = list(dict.fromkeys(v))
        v.sort()
        if len(v) == 2 and v not in combine:
            combine.append(v)
data = pd.DataFrame(combine)
data.to_excel('combine_2.xlsx', sheet_name='combine_2')
```
