```python
import pandas as pd
import os
import qrcode
from PIL import Image
from PIL import ImageDraw
from PIL import ImageFont
from qrcode.main import QRCode
df2 = pd.DataFrame(pd.read_csv('df1.csv'))
for line in df2['New Allocation']:
    qr = qrcode.QRCode(
           version=2,
           error_correction=qrcode.constants.ERROR_CORRECT_L,
           box_size=10,
           border=4,)
    qr.add_data(line)
    img = qr.make_image()
    draw = ImageDraw.Draw(img)
    font = ImageFont.truetype("arialbd.ttf",40)

    draw.text((45,290),line,font=font)

    imagepath = os.path.join('img', line+".png")
    img.save(imagepath) # Save image.
```
