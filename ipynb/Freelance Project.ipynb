```python
import pandas as pd
import sqlite3
```

```python
conn = sqlite3.connect('Umar_data.db')
cur = conn.cursor()
```

```python
sql ='''SELECT id, symbol, `datetime`, open, high, low, close, volume, vwap, num_trades, atr, adtv,
 ema10, sma50, sma100, sma200, uptrend, downtrend, candle, daily_gain,
(SELECT AVG(daily_gain)
FROM (SELECT daily_gain
FROM "alpaca_daily"
ORDER BY daily_gain
LIMIT 2
OFFSET (SELECT (COUNT(*) - 1) / 2
FROM "alpaca_daily"))) as daily_median_gain,
(daily_gain - (SELECT AVG(daily_gain)
FROM (SELECT daily_gain
FROM "alpaca_daily"
ORDER BY daily_gain
LIMIT 2
OFFSET (SELECT (COUNT(*) - 1) / 2
FROM "alpaca_daily")))) as daily_excess_gain FROM "alpaca_daily";'''
cur.execute(sql)
```

```python
result = cur.fetchall()
data = pd.DataFrame(result)
data
```

```python
data.to_excel('Umar.xlsx')
```
