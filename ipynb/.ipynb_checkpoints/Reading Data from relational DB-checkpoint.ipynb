```python
import pandas as pd
```

```python
import sqlite3
```

In order to work with SQLite database from Python, we first hae to connect to it. We can do that using connect function, which return **connection** object

```python
conn = sqlite3.connect('chinook.db') # If the database exists then put it into the directory file or you have to create new database with your data.
```

Once we have a **Connection** object, we can then create a **Cursor** object. **Cursors** allow us to execut SQL command against database.

```python
cur = conn.cursor()
```

The **cursor** created has a method **execute**, which will receive SQL parameters to run against the database

The code below will fetch first 5 rows from the employees table

```python
cur.execute('SELECT * FROM employees LIMIT 5')
```

After execute the query, you can't assign the result to a variable, because we need to run another command to actually fetch the result

Using **fetchall** method to getch all of the results of the query

```python
results = cur.fetchall()
results
```

After get the result, we use pandas to put the result into DataFrame that processing easier.

```python
df = pd.DataFrame(results)
df
```

```python
df.head(1)
```

Close **connection** and **cursor** after work done to prevent sqlite database from being locked. It will raise an error if you want to update or do operation inside the databse.

```python
cur.close()
conn.close()
```

### Using pandas **read_sql** method

We can use the pandas **read_sql** function to read the results of a SQL query directly into a pandas DataFrame. The code will execute the query and it will return a DataFrame immediately. It has several advantage:
* It doesn't require us to create a **Cursor** object or call **fetchall** at the end
* It automatically reads in the names of the headers from the table
* It crates a DataFrame, so we can quickly explore the data.

```python
import pandas as pd
import sqlite3
```

```python
conn = sqlite3.connect('chinook.db')
```

```python
df = pd.read_sql('select * from employees;', conn)
```

```python
df.head(1)
```

```python
df = pd.read_sql('select * from employees;', conn, index_col = 'EmployeeId', parse_dates = ['BirthDate','HireDate'])
```

```python
df.head(1)
```

```python
df.info()
```

Do some operations with DataFrame

```python
df['ReportsTo'].isna().sum()
```

```python
df['ReportsTo'].mean()
```

```python
df['ReportsTo'] > 1.75
```

```python
df['City'] = df['City'].astype('category')
```

```python
df.info()
```

### Using pandas **read_sql_query** method

We can get the same result using **read_sql_query** method

### Using **read_sql_table** method

**read_sql_table** is a useful function, it works only with SQLAlchemy, a Python SLQ toolkit and Object Relational Mapper

This is just a demonstration of its usage where we read the whole employees table.

```python
# first install SQLAlchemy 
import pandas as pd
import sqlite3
from sqlalchemy import create_engine
```

```python
engine = create_engine('sqlite:///chinook.db') # in case , SQL don't work, you have to specify exactly the path of db files.
connection = engine.connect()
```

# If there exist issues when connect to database, let's install **jupyerlab-sql**

```python
df=pd.read_sql_table('employees', con=connection)
```

```python
df
```

```python
df= pd.read_sql_table('employees', con =connection, index_col = 'EmployeeId', parse_dates = ['BirthDate', 'HireDate'])
```

```python
df.head(2)
```

```python
df.info()
```

```python
connection.close()
```

### Create tables from DataFrame Objects

Finally, we can persist DataFrame objects we've looking on in a database using to_sql method.

Although it is easy to implement it, an it could be a very slow process

```python
df.head(2)
```

```python
df.to_sql?
```

```python
conn = sqlite3.connect('chinook.db')
cur= conn.cursor()
```

```python
cur.execute('drop table if exists employee2;')
df.to_sql('employee2', conn)
```

```python
cur.execute('select * from employee2')
df = cur.fetchall()
df = pd.DataFrame(df)
df
```

```python
# or a shorten way is :
pd.read_sql('select * from employee2', conn).head()
```
