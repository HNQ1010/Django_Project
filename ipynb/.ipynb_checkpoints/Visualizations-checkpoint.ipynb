```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Global API

Matplotlib's default global API has a global, MATLAB-style interface

```python
x = np.arange(-10, 11)
```

```python
plt.figure(figsize = (8,4))
plt.plot(x, x**2)
plt.plot(x, -1*(x**2))
plt.title('Eg Plot')
```

```python
plt.figure(figsize = (12,6))
plt.title('Eg2 Plot')
plt.subplot(1,2,1) # rows, columns, and index . 1<=index<=2
plt.plot(x, x+1) # plot 1
plt.plot([0,0,0], [-10, 0, 100]) # plot 2 [0,-10], [0,0], [0,100]
plt.legend(['X^2', 'Vertical Line']) # 'X^2' for (x,x**2) and 'Vertical Line' for ([0,0,0],[-10, 0 ,100])
plt.xlabel('X')         # x label
plt.ylabel('X + 1') # y label

plt.subplot(1,2,2)
plt.plot(x, x+2)
plt.plot([-10, 0, 10],[-50, -20, -50])
plt.legend(['-X^2, Horizontal Line'])
plt.xlabel('X')         # x label
plt.ylabel('X + 2') # y label
```

### OOP Interface

```python
fig, axes = plt.subplots(figsize=(8,4)) # use subplots instead of subplot(global API)
```

```python
axes.plot(x, (x**2), color='red', linewidth =3, marker ='o', markersize = 8, label ='X^2')
axes.plot(x, -1*(x**2), 'b--', label = '-X^2') # 'b--' color and marker
axes.set_xlabel('X')
axes.set_ylabel('X Squared')
axes.legend()
axes.set_title('Plot')
fig
```

```python
fig, axes = plt.subplots(figsize = (8,4))

axes.plot(x, x+0, linestyle = 'solid')
axes.plot(x, x+1, linestyle = 'dashed')
axes.plot(x, x+2, linestyle = 'dashdot')
axes.plot(x, x+3, linestyle = 'dotted');
axes.set_title('Colour plot')          
```

```python
fig, axes = plt.subplots(figsize = (8,4))

axes.plot(x, x+0, '--og', label = 'solid green')
axes.plot(x, x+1, '--c', label= 'dashcyan')
axes.plot(x, x+2, '-.b', label= 'dashdot blue')
axes.plot(x, x+3, ':r', label= 'dotted red')
axes.set_title('Other Colour Plot')
axes.legend()
```

### Other types of plots

#### Figure and subfigure

When we call the **subplots()** function we get a tuple containing a **figure** and **axes** element.

```python
plot_object = plt.subplots()
fig, ax = plot_object
ax.plot([1, 2, 3],[1, 2, 3])
plot_object
```

We can also define how many elements we want inside our figure. To do that we can set the **nrows** and **ncols** params.

```python
plot_object = plt.subplots(nrows =2, ncols = 2, figsize = (8,4))
fig, ((ax1, ax2), (ax3, ax4)) = plot_object
#plot_object
ax1.plot(np.random.randn(50), c='red', linestyle = '--')
ax2.plot(np.random.randn(50), c='green', linestyle = ':')
ax3.plot(np.random.randn(50), c='blue', marker= 'o', linewidth = 3.0)
ax4.plot(np.random.randn(50), c='yellow')
```

```python
plot_object = plt.subplots(nrows =2, ncols = 1, figsize = (8,4))
fig, (ax1,ax2) = plot_object
#plot_object
ax1.plot(np.random.randn(50), c='red', linestyle = '--')
ax2.plot(np.random.randn(50), c='green', linestyle = ':')
```

#### The **subplot2grid** command

There is another way to make subplots using grid-like format:
**subplot2grid(shape, loc, rowspan, columnspan, *kwarg)**
* shape: shape of grid in which to place axis, first entry is **number of rows, second is number of cols**
* loc: location to place axis within the grid. First is **row number, second is column number**
* rowspan: expand row
* columnspan: expand column
* kwarg:...

```python
plt.figure(figsize = (12,6))
ax1=plt.subplot2grid((3,3), (0,0), colspan =3).plot(np.random.randn(50), c='red', linestyle = '--')
ax2=plt.subplot2grid((3,3), (1,0), colspan =2).plot(np.random.randn(50), c='red', linestyle = '--')
ax3=plt.subplot2grid((3,3),(1,2), rowspan =2).plot(np.random.randn(50), c='green', linestyle = ':')
ax4=plt.subplot2grid((3,3),(2,0)).plot(np.random.randn(50), c='blue', marker= 'o', linewidth = 3.0)
ax5=plt.subplot2grid((3,3),(2,1)).plot(np.random.randn(50), c='yellow')
```

```python
plt.figure(figsize = (20,10))
ax1=plt.subplot2grid((4,4), (0,0)).plot(np.random.randn(50), c='black', linestyle = '--')
ax2=plt.subplot2grid((4,4), (0,1)).plot(np.random.randn(50), c='red', linestyle = '--')
ax3=plt.subplot2grid((4,4), (0,2)).plot(np.random.randn(50), c='green', linestyle = '--')
ax4=plt.subplot2grid((4,4), (0,3)).plot(np.random.randn(50), c='blue', linestyle = '--')
ax5=plt.subplot2grid((4,4), (1,0)).plot(np.random.randn(50), c='yellow', linestyle = '--')
```

### Scatter Plot

```python
N =50
x =np.random.rand(N)
y =np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (20*np.random.rand(N))**2 # 0 to 15 point radii
```

```python
plt.figure(figsize = (12,6))
plt.scatter(x, y, s=area, c=colors, alpha=0.5, cmap = 'Spectral')
plt.colorbar()
plt.show()
```

```python
fig = plt.figure(figsize = (12,6))
ax1 = fig.add_subplot(1,2,1)
plt.scatter(x, y, s=area, c=colors, alpha =0.5, cmap='Pastel1')
plt.colorbar()
ax2= fig.add_subplot(1,2,2)
plt.scatter(x, y, s=area, c=colors, alpha=0.5, cmap='Pastel2')
plt.colorbar()
```

Full cmap(colormap) options at https://matplotlib.org/users/colormaps.html

### Histograms

```python
values = np.random.randn(1000)
plt.subplots(figsize = (12,6))
plt.hist(values, bins=100, alpha=0.8, histtype ='bar', color= 'steelblue', edgecolor ='green')
plt.xlim(xmin = -5, xmax=5)
plt.show()
```

### KDE (kernel density estimation)

```python
from scipy import stats
```

```python
values = np.random.randn(1000)
```

```python
density = stats.kde.gaussian_kde(values)
density

plt.subplots(figsize =(12,6))
value2 = np.linspace(min(values)-10, max(values) +10, 100)
plt.plot(value2, density(value2), color ='#FF7F00')
plt.fill_between(value2, 0, density(value2), alpha=0.5, color='#FF7F00')
plt.xlim(xmin=-5, xmax=5)
plt.show()
```

### Combine plots

```python
plt.subplots(figsize = (12,6))
plt.hist(values, bins =100, alpha=0.8, density=1, histtype='bar', color='steelblue', edgecolor ='green')
plt.plot(value2, density(value2), color='#FF7F00', linewidth=3.0)
plt.xlim(xmin=-5, xmax=5)
plt.show()
```

### Bar plots

```python
Y = np.random.rand(1,5)[0]
Y2 = np.random.rand(1,5)[0]
```

```python
plt.figure(figsize=(12,6))
barWidth = 0.5
plt.bar(np.arange(len(Y)), Y, width=barWidth, color='#00b894')
plt.show()
```

Also can be stacked bars, and add a legend to the plot:

```python
plt.figure(figsize= (12,6))
barWidth=0.5
plt.bar(np.arange(len(Y)), Y, width = barWidth, color='#00b894', label = 'Label Y')
plt.bar(np.arange(len(Y2)), Y2, width = barWidth, color='#e17055', bottom=Y,label ='Label Y2')
plt.legend()
plt.show()
```
