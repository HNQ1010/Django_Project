```python
def mklbl(prefix, n):
    return ['%s%s' %(prefix, i) for i in range(n)]
miindex = pd.MultiIndex.from_product([mklbl('A', 4), mklbl('B', 2), mklbl('C', 4), mklbl('D', 2)])
micolumns = pd.MultiIndex.from_tuples([('a', 'foo'), ('a', 'bar'), ('b', 'foo'), ('b', 'bah')], names=['lvl_0', 'lvl_1'])
df = (pd.DataFrame(np.arange(len(miindex) * len(micolumns)).reshape(len(miindex), len(micolumns)), index=miindex, columns=micolumns)).sort_index().sort_index(axis=1)
```
